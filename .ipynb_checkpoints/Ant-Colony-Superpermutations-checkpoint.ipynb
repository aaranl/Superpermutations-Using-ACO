# Ant Colony Superpermutations Solution

## Super-permutations

The problem of superpermutations is a interesting topic, firstly what is a super-permutation, a super-permutation can be explained as a string that contains each permutation of $n $ symbols. But, what we want to figure out is the most optimal and shortest superpermutation, because under that defintion you could have something like $ n = 2 $ and then just concatenate the possible permutations of $ 1-2 $ and $ 2-1 $ making $ 1-2-2-1 $ this technically is a superpermutation but this is called a trivial superpermutation.

What we want to figure out is the shortest super-permutation, but also the most optimal, shortest is easy to figure out, but what does it mean to be optimal? 

This isn't an easy question to answer because it would all determine on your specific use case, but I can help conceptualize what could be optimal. In the case of n > 5 there are many possible solutions that mathmatically meet the terms of "optimal" meaning they are the shortest possible super-permutations in terms of length. 

So lets go through a thought experiment, lets say you had 6 features, that can be performed in any order, but order can have affect of the overall efficiency. To be able to test this optimally, you would use a super-permutation to have the most efficient testing cycles. There are going to be many different super-permutations at this level in which you could apply, but they won't give the same results. This is the problem with super-permutations above $ n >= 5 $ there is no real way to figure out the specific super-permutation for your use case, so how do you find it?

I don't know yet, but my idea is to apply an ant colony optimization technique which has been applied to problems like the traveling salesman. Using this technique to find super-permutations is already a problem within itself but the abstraction of using an ant colony is required to modify it for your use case. If a specific permutation of your features would give a far better outcome than another permutation, there could be a superpermutation that uses that permutation far more. Using ants you could find this super-permutation, atleast thats the idea. 

To do this the ants would go between nodes, and you could give more weight to specific permutations that worked "better". This is the very dumbed down explanation of what I want to attempt to do. 

## Part One - Traveling Salesmen Problem using Ant Colony Optimization

What we are trying to do is implement an Ant Colony Optimization (ACO) algorithm. This has already been done for problems like the traveling sales person (TSP). So for my first implementation I want to try and do this exact implementation. I believe it is important to understand how the ACO works in regards to TSP before I can began to optimize it for my case of finding superpermutations. 


```python
# (1)
def distance(location1, location2):
    return np.sqrt(np.sum((location1-location2)**2))

# (2)
def ant_colony_optimization(points, n_ants, n_iterations, pheromone_influence, distance_influence, evaporation_rate, pheromones_layed):
    n_points = len(points)
    pheromone_matrix = np.ones((n_points, n_points)) # (3)
    best_path = None
    best_path_length = np.inf # (4)
    
    for curr_iteration in range(n_iterations):
        list_of_paths = []
        list_of_path_lengths = []
        
        for ant in range(n_ants):
            visited = n_points * [False] # (5)
            curr_point = np.random.randint(n_points)
            visited[curr_point] = True
            path = [curr_point] # (6)
            path_length = 0 
            
            while False in visited: 
                unvisited = np.where(np.logical_not(visited))[0] # (7)
                probabilities = np.zeros(len(unvisited))

                
                for i, unvisited_point in enumerate(unvisited):
                    # (8)
                    probabilities[i] = pheromone_matrix[curr_point, unvisited_point]**pheromone_influence / distance(points[curr_point], points[unvisited_point])**distance_influence
                    
                probabilities /= np.sum(probabilities) # (9)
                
                next_point = np.random.choice(unvisited, p=probabilities) # (10)
                path.append(next_point)
                visited[next_point] = True
                path_length += distance(points[curr_point], points[next_point])
                curr_point = next_point
                
            list_of_paths.append(path)
            list_of_path_lengths.append(path_length)
            
            if path_length < best_path_length:
                best_path = path
                best_path_length = path_length
        
            
        pheromone_matrix *= evaporation_rate #(11)
        
        for path, path_length in zip(list_of_paths, list_of_path_lengths):
            for i in range(n_points - 1):
                # (12)
                pheromone_matrix[path[i], path[i+1]] += pheromones_layed / path_length
            # (13)
            pheromone_matrix[path[-1], path[0]] += pheromones_layed / path_length
    return list_of_paths, best_path


# (14)
def visualize_paths_2d(points, paths, best_path):
    plt.figure(figsize=(10, 8))
    plt.scatter(points[:, 0], points[:, 1], c='blue')

    for path in paths:
        for i in range(len(path) - 1):
            plt.plot([points[path[i], 0], points[path[i+1], 0]], [points[path[i], 1], points[path[i+1], 1]], 'grey', alpha=0.5)

    if best_path is not None:
        for i in range(len(best_path) - 1):
            plt.plot([points[best_path[i], 0], points[best_path[i+1], 0]], [points[best_path[i], 1], points[best_path[i+1], 1]], 'red', linewidth=2)

    plt.xlabel('X coordinate')
    plt.ylabel('Y coordinate')
    plt.title('Visualization of Ant Colony Paths')
    plt.show()

points = np.random.rand(50,2)         
paths, best_path = ant_colony_optimization(points, n_ants=500, n_iterations=50, pheromone_influence=1, distance_influence=1, evaporation_rate=0.5, pheromones_layed=1)
visualize_paths_2d(points, paths, best_path)

            
 
    
        
```

### Explanation


##### Distance Formula

The first function we need is the ability to calculate distance, and the most proper way to do this is math! We do this through  the Euclidean distance formula, 

$$ d(P, Q) = \sqrt{(p_1 - q_1)^2 + (p_2 - q_2)^2 + \ldots + (p_n - q_n)^2} $$

which in our case since we are only working with two points would be,

$$ d(x, y) = \sqrt{(x_1 - y_1)^2 + (x_2 - y_2)^2} $$

translate this into code and we get our distance function at : $(1)$

Moving on we get into the actual way we are going to figure out this problem at : $(2)$ : This is a very long function but lets directly lay out what each attribute correlates to.


#### Ant Colony Optimization Construction

###### Parameters:

- `points` (array_like): 
    An array of points representing the locations to be connected. Each point should be an n-dimensional tuple or list.

- `n_ants` (int): 
    The number of ants used in the simulation. Each ant independently explores possible paths between points, contributing to the optimization process.

- `n_iterations` (int): 
    The number of runs for the algorithm. Each run consists of all ants completing a path and updating pheromone trails. More runs allow for more refinement of the optimal path.

- `pheromone_influence` (float): 
    This parametercontrols the influence of the pheromone trails on the path selection by ants. A higher value increases the tendency of ants to follow stronger pheromone trails.

- `distance_influence` (float): 
    This parameter controls the influence of the heuristic information (i.e., distance). A higher value makes ants more likely to choose shorter paths, prioritizing closeness.

- `evaporation_rate` (float): 
    The rate at which pheromones evaporate. This is a factor between 0 and 1. Pheromone evaporation prevents the convergence to a locally optimal path and encourages exploration of new paths.

- `pheromones_layed` (float): 
    Represents the quantity of pheromones laid by ants on a path. It influences how much the path is reinforced for future ants. A higher value means more pheromone is left on shorter paths, encouraging future ants to follow these paths.

###### Returns

- `best_path` (list): 
    The best path found by the ants, represented as a list of point indices in the order they are connected.

- `best_path_length` (float): 
    The length of the best path found.



#### Line-by-Line Explanations

I am now going to go over specific lines in the ant_colony_optimization function that I believe may need some understanding if you are not a big user of NumPy or matrixes in general.

- $(3)$ : `pheromone_matrix = np.ones((n_points, n_points))` : The numpy ones function creates an array where all elements are initialized to 1. So in regards to our implementation the pheromone matrix represents the amount of pheromone on the paths between points. Initially, you want to give an equal chance for any path to be chosen, so you start with a matrix where all paths have the same amount of pheromone. This is why it's initialized to a matrix of ones.


- $(4)$ : `best_path_len = np.inf` : This one is quite simple but to explain, np.inf is infinity, and best_path_len is supposed to be the shortest possible length. So to ensure and avoid many problems we first initiate it to inf so when we wnat to replace it later, it is guaranteed to be replaced because nothing is bigger than inf. 

- $(5)$ :`visited = n_points * [False]` : Again quite a simple line, but to explain the visited list is being used to track whether each of these points has been visited. Initially, all elements are set to False, indicating that none of the points have been visited yet.

- $(6)$ : `path = [curr_point]` : This may seem like a very simple line of code but this is setting up a foundational element of the algorithm  the path that starts at the initial point. This path will be built upon and modified as the algorithm runs, and is a key component in finding an optimal solution to the problem at hand.

- $(7)$ : `unvisited_indexes = np.where(np.logical_not(visited))[0]` : This one is a bit weird but lets explain step-by-step, np.logical_not is going to take the list of boolean values in visited and just flip every value, so true becomes false and false becomes true. Now we pass that into np.where which is going to return the indices of all the values that are true, which are the unvisited locations. Finally, the [0] is just to fix the issue where the np.where function returns a tuple and we just need to access the first element in that tuple. 

- $(8)$ : `probabilities[i] = (pheromone_matrix[curr_point, unvisited_point]**pheromone_influence) / (distance(points[curr_point], points[unvisited_point])**distance_influence)` : This one is a doozy, but lets go over it first we have the phermone_matrix, this matrix is based off of an earlier entry where we set up a matrix that would be $ n x n $ size based on how many points there are. What this first part does is essentially take your curr_point or location in that matrix, relates it to a unvisited_point somewhere else in that matrix, and calculates how much pheromone_influence you would have to go to that point. Moving on to the second portion of this we are calculating distance and the reason for distance_influnce. The first portion is easy, there is a distance between curr_point, and unvisited_point, what is it. But why do we raise it to power of distance_influence. The reason for this is simple, we want higher distances to have more of a influence on the system so that ants can be more inclined to either choose a high pheromone level path, or a short path. Both of these working together is how an ant can determine what direction to go in algorithmicly so thats why we find the average of the numbers by dividing and this is how our ant will make decisions. 

- $(9)$ : `probabilities /= np.sum(probabilities)` : For this the probabilites array is a collection of probabilities but to be able to work with this easier, its better to normalize the array. To do this we do some easy division of each element by the sum of all the probabilites in the array. This ensures that they will add up to 1.0. In a mathematic sense this now means each entry is now the chance that we will choose that entry which is much easier to work with especially with what we are going to do next.

- $(10)$ : `next_point = np.random.choice(unvisited, p=probabilities)` : This line is simple, but requires some explaining because of the NumPy library. Essentially what is happening is we are randomly selecting one point from unvisited, and related to its probability value and making a choice for the next_point we will visit.

- $(11)$ : `pheromome_matrix *= evaporation_rate` : This is going to modify our entire pheromone_matrix, by our evaporation_rate, which should be from 0.1-0.999..., this will ensure that our pheromones are will dissapate and is important to simulate how an ant acts.

- $(12)$ : `pheromone_matrix[path[i], path[i+1]] += pheromones_layed / path_length` : This is a crucial aspect of the algorithm, what is going to happen is, we are going to increase the pheromone levels on the path this ant took, which will be scaled by the amount of pheromones we are laying, and then to calculate for TSP, it will be divided in the length is further. This incentivizes ants to look for shorter paths. 

- $(13)$ : `pheromone_matrix[path[-1], path[0]] += pheromones_layed / path_length` : This is how we "return" with an ant, path[-1] is the last point in the path, and path[0] is obviously where we started. So after an ant has gone to all points, it is going to return back to the start, and increase the pheromone levels on that path like we did before.

## Idea Two Implementation

```python
import itertools as it
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from tqdm import tqdm

def generate_Permutations(n):
    n += 1
    permutations = it.permutations(range(1, n))
    permutations_as_strings = [''.join(map(str, perm)) for perm in permutations]
    return permutations_as_strings

def calculate_distance(perm_1, perm_2):
    perm_length = len(perm_1)
    distance = perm_length
    for offset in range(perm_length):
        if hamming_distance(perm_1[offset:], perm_2[:perm_length - offset]) == 0:
            distance = offset
            break
    return distance

def hamming_distance(s_1, s_2):
    return sum((char_1 != char_2) for char_1, char_2 in zip(s_1, s_2))

def distance(perm1, perm2):
    return calculate_distance(perm1, perm2)

def ant_colony_optimization(permutations, n_ants, n_iterations, pheromone_influence, distance_influence, evaporation_rate, pheromones_layed, seed=None):
    if seed is not None:
        np.random.seed(seed)
    n_perms = len(permutations)
    pheromone_matrix = np.ones((n_perms, n_perms))
    best_path = None
    best_path_length = np.inf

    with tqdm(total=n_iterations * n_ants, desc="Progress", position=0, leave=True) as pbar:
        for curr_iteration in range(n_iterations):
            list_of_paths = []
            list_of_path_lengths = []
            
            for ant in range(n_ants):
                visited = n_perms * [False]
                curr_perm = np.random.randint(n_perms)
                visited[curr_perm] = True
                path = [curr_perm]
                path_length = 0 
                
                while False in visited:
                    unvisited = np.where(np.logical_not(visited))[0]
                    probabilities = np.zeros(len(unvisited))
                    
                    for i, unvisited_perm in enumerate(unvisited):
                        probabilities[i] = pheromone_matrix[curr_perm, unvisited_perm]**pheromone_influence / distance(permutations[curr_perm], permutations[unvisited_perm])**distance_influence
                        
                    probabilities /= np.sum(probabilities)
                    
                    next_perm = np.random.choice(unvisited, p=probabilities)
                    path.append(next_perm)
                    visited[next_perm] = True
                    path_length += distance(permutations[curr_perm], permutations[next_perm])
                    curr_perm = next_perm
                    
                list_of_paths.append(path)
                list_of_path_lengths.append(path_length)
                
                if path_length < best_path_length:
                    best_path = path
                    best_path_length = path_length

                pbar.update(1)
                pbar.set_postfix(iter=f"{curr_iteration+1}/{n_iterations}", 
                                 ant=f"{ant+1}/{n_ants}")
            
            pheromone_matrix *= evaporation_rate
            
            for path, path_length in zip(list_of_paths, list_of_path_lengths):
                for i in range(n_perms - 1):
                    pheromone_matrix[path[i], path[i+1]] += pheromones_layed / path_length
                pheromone_matrix[path[-1], path[0]] += pheromones_layed / path_length
    
    return list_of_paths, best_path

def visualize_paths(permutations, paths, best_path):
    G = nx.Graph()
    for i, perm in enumerate(permutations):
        G.add_node(i, label=perm)
    
    plt.figure(figsize=(12, 10))
    pos = nx.spring_layout(G, k=0.5, iterations=50)
    
    for path in paths:
        for i in range(len(path) - 1):
            G.add_edge(path[i], path[i+1], color='lightgray', alpha=0.5)
    
    if best_path is not None:
        for i in range(len(best_path) - 1):
            G.add_edge(best_path[i], best_path[i+1], color='r', width=2)

    nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=3000)
    
    edge_colors = [G[u][v].get('color', 'lightgray') for u, v in G.edges()]
    edge_widths = [G[u][v].get('width', 1) for u, v in G.edges()]
    nx.draw_networkx_edges(G, pos, edge_color=edge_colors, width=edge_widths)
    
    labels = {i: perm for i, perm in enumerate(permutations)}
    nx.draw_networkx_labels(G, pos, labels, font_size=10, font_weight='bold')
    
    plt.title('Visualization of Ant Colony Paths on Permutations')
    plt.axis('off')
    plt.tight_layout()
    plt.show()

def construct_superpermutation(permutations, best_path):
    result = permutations[best_path[0]]
    for i in range(1, len(best_path)):
        prev_perm = permutations[best_path[i-1]]
        curr_perm = permutations[best_path[i]]
        overlap = len(prev_perm) - calculate_distance(prev_perm, curr_perm)
        result += curr_perm[overlap:]
    return result

def analyze_superpermutation(permutations, best_path):
    superperm = construct_superpermutation(permutations, best_path)
    
    print(f"Superpermutation: {superperm}")
    print(f"Length: {len(superperm)}")
    
    all_present = all(perm in superperm for perm in permutations)
    print(f"All permutations present: {all_present}")
    
    print("\nPath of permutations:")
    for index in best_path:
        print(permutations[index], end=" -> ")
    print(permutations[best_path[0]])
    
    return superperm

n_ants = 20
n_iterations = 10
pheromone_influence = 0.7
distance_influence = 10
evaporation_rate = 0.3
pheromones_layed = 1
seed = 42
n = 5
visualize = True
permutations = generate_Permutations(n)

paths, best_path = ant_colony_optimization(permutations,n_ants,n_iterations,pheromone_influence,distance_influence,evaporation_rate,pheromones_layed,seed)
if visualize is True:
    visualize_paths(permutations, paths, best_path)
superpermutation = analyze_superpermutation(permutations, best_path)
```

The formula for the lower bound for [superpermutations](https://en.wikipedia.org/wiki/Superpermutation) was found in a very interesting way and I recommend researching into the Haruhi Problem on the link. The formula as is below:

$$ n! + (n1)! + (n2)! + n  3 $$

Now lets go ahead and plug in 5 to get an idea of how my algorithm performed in finding the lowest possible superpermutation to see how we performed.

$$ (5)! + ((5)1)! + ((5)2)! + (5)  3 $$

$$ 120 + 24 + 6 + 2 = 152 $$

Keep in mind the formula just provides a lower bound which means that the length of the shortest supermutation cannot be less than 152, but it does not guarantee that 152 is achievable. So in the case for n=5, the lower bound is not actually achievable at 152, but my implementation of this algorithm correctectly finds either the shortest achievable or gets very close.
